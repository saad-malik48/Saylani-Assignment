```python
import math
import cmath  

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

discriminant = b**2 - 4*a*c

if discriminant >= 0:
  
    x1 = (-b + math.sqrt(discriminant)) / (2*a)
    x2 = (-b - math.sqrt(discriminant)) / (2*a)
    print("Root 1 =", x1)
    print("Root 2 =", x2)
else:
   
    x1 = (-b + cmath.sqrt(discriminant)) / (2*a)
    x2 = (-b - cmath.sqrt(discriminant)) / (2*a)
    print("Root 1 =", x1)
    print("Root 2 =", x2)
```

```python
#18) Future Value in Savings:
# Write a Python program to calculate future value of investment:
# FV = PV(1 + r)^t
```

```python
import decimal
decimal.getcontext().prec = 28
PV = decimal.Decimal(input("Present value: "))
r = decimal.Decimal(input("Annual Rate(decimal): "))  
t = decimal.Decimal(input("Years: "))
FV = PV * (1 + r) ** t

if FV > 10**20:
    print("Future value= {:.2e}".format(FV))  
else:
    print("Future value=", float(FV))  
```

```python
#19) Work Done by a Force:
#Write a Python program to calculate work done using:
#W = f \cdot d \cdot \cos(\theta)
```

```python
import math

f = float(input("Force: "))
d = float(input("Distance: "))
theta = float(input("Angle in degrees: "))

W = f * d * math.cos(math.radians(theta))
print("Work Done =", W)
```
